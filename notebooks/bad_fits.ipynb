```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
from pathlib import Path
import seaborn as sns
import numpy as np
import pandas as pd
import json
sns.reset_orig()
from tqdm.notebook import tqdm
import shutil
```

```python
pickdir1 = 'bad_fits'
pickdir2 = 'good_fits'
```

```python
outdir = 'outdir_b_all_efac100'
cmd = "C:\\Python311\\python.exe fit.py --outdir outdir_b_all_efac100 --gaussian --efac 100."
var = 'log_evidence' if outdir.startswith('outdir_b') else 'aic'
with open(f'../output/results_{var}_from_{outdir}.json', 'r', encoding='utf-8-sig') as f:
    results = pd.DataFrame(json.load(f)).T
results['best_fitting'] = results.iloc[:, :].idxmax(axis=1) if var == 'log_evidence' else results.iloc[:, :].idxmin(axis=1)
```

```python
Path(f'../output/{outdir}/{pickdir1}').mkdir(parents=True, exist_ok=True)
for file in os.listdir(f'../output/{outdir}/{pickdir1}'):
    os.remove(f'../output/{outdir}/{pickdir1}/{file}')
Path(f'../output/{outdir}/{pickdir2}').mkdir(parents=True, exist_ok=True)
for file in os.listdir(f'../output/{outdir}/{pickdir2}'):
    os.remove(f'../output/{outdir}/{pickdir2}/{file}')
    
if not os.path.exists(f'../output/{outdir}/bad_fits.txt'):
    with open(f'../output/{outdir}/bad_fits.txt', 'w', encoding='utf-8-sig') as f:
        f.write('')
bad_fits = open(f'../output/{outdir}/bad_fits.txt', 'r', encoding='utf-8-sig').read().strip().splitlines()
for jname in tqdm(results.index):
    if not os.path.exists(f'../output/{outdir}/{jname}/{results.loc[jname, "best_fitting"]}_result.png'):
        print(f'../output/{outdir}/{jname}/{results.loc[jname, "best_fitting"]}_result.png' + " not exist")
        os.system(f"cd D:\\spectral_fit & {cmd} --jname {jname} --model {results.loc[jname, 'best_fitting']}")

    if jname in bad_fits:
        shutil.copy(f'../output/{outdir}/{jname}/{results.loc[jname, "best_fitting"]}_result.png', f'../output/{outdir}/{pickdir1}/{jname}.png')
    else:
        shutil.copy(f'../output/{outdir}/{jname}/{results.loc[jname, "best_fitting"]}_result.png', f'../output/{outdir}/{pickdir2}/{jname}.png')

# open the two folders externally
os.startfile(os.path.abspath(f'../output/{outdir}/{pickdir1}'))
os.startfile(os.path.abspath(f'../output/{outdir}/{pickdir2}'))
```

```python
# read bad-fit jnames from the directory instead of the text file
bad_fits = list(map(lambda x: x.removesuffix('.png'), os.listdir(f'../output/{outdir}/{pickdir1}')))

# save the bad fits to a text file
with open(f'../output/{outdir}/bad_fits.txt', 'w', encoding='utf-8-sig') as f:
    f.write('\n'.join(bad_fits))
```

```python
pickdir3 = 'msps'
pickdir3_content = open(f'../catalogue/{pickdir3}.txt', 'r', encoding='utf-8-sig').read().strip().splitlines()
pickdir_gps = 'pickdir3_content'
Path(f'../output/{outdir}/{pickdir_gps}').mkdir(parents=True, exist_ok=True)
for file in os.listdir(f'../output/{outdir}/{pickdir_gps}'):
    os.remove(f'../output/{outdir}/{pickdir_gps}/{file}')
for jname in tqdm(pickdir3_content):
    shutil.copy(f'../output/{outdir}/{jname}/{results.loc[jname, "best_fitting"]}_result.png', f'../output/{outdir}/{pickdir_gps}/{jname}.png')
    
os.startfile(os.path.abspath(f'../output/{outdir}/{pickdir_gps}'))
```
